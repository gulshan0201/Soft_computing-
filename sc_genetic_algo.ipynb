```python
import random
```


# Parameters


```python
POP_SIZE = 4
CHROM_LEN = 5
X_MIN = 0
X_MAX = 2
MUTATION_RATE = 0.1
```


# Binary → Decimal


```python

def binary_to_decimal(chrom):
    decimal = int(chrom, 2)
    x = X_MIN + decimal * (X_MAX - X_MIN) / (2**CHROM_LEN - 1)
    return x
```


# Fitness Function
# f(x) = -x^2 + 2x

```python
def fitness(x):
    return -x**2 + 2*x
```



# Initialize Population

```python
def init_population():
    return [''.join(random.choice('01') for _ in range(CHROM_LEN))
            for _ in range(POP_SIZE)]
```


# Roulette Wheel Selection

```python
def roulette_selection(population):
    fitness_values = [fitness(binary_to_decimal(c)) for c in population]
    total_fitness = sum(fitness_values)

    probs = [f / total_fitness for f in fitness_values]

    # cumulative probability
    cum_probs = []
    cum_sum = 0
    for p in probs:
        cum_sum += p
        cum_probs.append(cum_sum)

    # selection
    new_population = []
    for _ in range(POP_SIZE):
        r = random.random()
        for i in range(len(cum_probs)):
            if r <= cum_probs[i]:
                new_population.append(population[i])
                break

    return new_population
```



# One-Point Crossover (MIDPOINT)

```python
def crossover_mid(population):
    new_pop = []
    point = CHROM_LEN // 2   # fixed middle

    for i in range(0, POP_SIZE, 2):
        p1 = population[i]
        p2 = population[i+1]

        c1 = p1[:point] + p2[point:]
        c2 = p2[:point] + p1[point:]

        new_pop.extend([c1, c2])

    return new_pop
```


# Mutation (Bit Flip)

```python

def mutation_bit_flip(population):
    new_population = []

    for chrom in population:
        # Convert string to list
        chrom_list = list(chrom)

        # Select random position
        pos = random.randint(0, len(chrom_list) - 1)

        # Flip bit
        if chrom_list[pos] == '0':
            chrom_list[pos] = '1'
        else:
            chrom_list[pos] = '0'

        # Convert back to string
        new_population.append(''.join(chrom_list))

    return new_population
```

```python
population = init_population()
print("Initial Population:", population)

population = roulette_selection(population)
print("After Selection:", population)

population = crossover_mid(population)
print("After Crossover (Midpoint):", population)

population = mutation_bit_flip(population)
print("After Mutation:", population)

print("\nFinal Output:")
for chrom in population:
    x = binary_to_decimal(chrom)
    print(chrom, "-> x =", round(x, 2), "f(x) =", round(fitness(x), 2))
```
