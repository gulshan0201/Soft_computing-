```python
import numpy as np

#Data
X = np.array([
    [800, 2],
    [1000, 3],
    [1200, 3],
    [1500, 4],
    [1800, 4],
    [2000, 5],
    [2200, 5],
    [2500, 6],
    [2700, 6],
    [3000, 7]
])

# Target: Price (in lakhs)
y = np.array([
    30, 40, 50, 65, 80,
    95, 110, 130, 150, 180
]).reshape(-1, 1)


# NORMALIZATION 
X = (X - X.mean(axis=0)) / X.std(axis=0)


# NETWORK ARCHITECTURE
input_size = 2
hidden1_size = 6
hidden2_size = 4
output_size = 1


# INITIALIZE WEIGHTS
np.random.seed(42)

W1 = np.random.randn(input_size, hidden1_size)
b1 = np.zeros((1, hidden1_size))

W2 = np.random.randn(hidden1_size, hidden2_size)
b2 = np.zeros((1, hidden2_size))

W3 = np.random.randn(hidden2_size, output_size)
b3 = np.zeros((1, output_size))


# ACTIVATION FUNCTIONS
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)


# TRAINING SETTINGS
lr = 0.01
epochs = 500

losses = []


# TRAINING LOOP
for epoch in range(epochs):

    
    Z1 = np.dot(X, W1) + b1
    A1 = relu(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = relu(Z2)

    Z3 = np.dot(A2, W3) + b3
    y_hat = Z3   # Linear output (regression)

    
    loss = np.mean((y_hat - y) ** 2)
    losses.append(loss)

    
    dZ3 = (y_hat - y)

    dW3 = np.dot(A2.T, dZ3)
    db3 = np.sum(dZ3, axis=0, keepdims=True)

    dA2 = np.dot(dZ3, W3.T)
    dZ2 = dA2 * relu_derivative(Z2)

    dW2 = np.dot(A1.T, dZ2)
    db2 = np.sum(dZ2, axis=0, keepdims=True)

    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * relu_derivative(Z1)

    dW1 = np.dot(X.T, dZ1)
    db1 = np.sum(dZ1, axis=0, keepdims=True)

    # Update  
    W3 -= lr * dW3
    b3 -= lr * db3

    W2 -= lr * dW2
    b2 -= lr * db2

    W1 -= lr * dW1
    b1 -= lr * db1

    # Print Loss
    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# Preiction
print("\nFinal Predictions vs Actual:\n")
for i in range(len(X)):
    print(f"Predicted: {y_hat[i][0]:.2f}  -->  Actual: {y[i][0]}")

print("\nFinal Loss:", losses[-1])

def predict(size, bedrooms):
    new_data = np.array([[size, bedrooms]])
    new_data = (new_data - X.mean(axis=0)) / X.std(axis=0)

    Z1 = np.dot(new_data, W1) + b1
    A1 = relu(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = relu(Z2)

    Z3 = np.dot(A2, W3) + b3

    return Z3[0][0]

# prediction
print("\nNew Prediction (1600 sq ft, 3 bedrooms):")
print(f"Predicted Price: {predict(1600, 3):.2f} lakhs")
```
