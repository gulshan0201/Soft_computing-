UNION

```python
# Fuzzy Union

class FuzzySet:
    def __init__(self, membership):
        self.mu = membership

    def union(self, other):
        # Zadeh union (max)
        return {
            x: max(self.mu.get(x, 0), other.mu.get(x, 0))
            for x in set(self.mu) | set(other.mu)
        }


# Example
A = FuzzySet({
    'x1': 0.7,
    'x2': 0.4,
    'x3': 0.9
})

B = FuzzySet({
    'x1': 0.6,
    'x2': 0.8,
    'x3': 0.2
})

print("Fuzzy Set A:", A.mu)
print("Fuzzy Set B:", B.mu)
print("Union (A ∪ B):", A.union(B))
```

Fuzzy Intersection (AND)

```python
class FuzzySet:
    def __init__(self, membership):
        self.mu = membership

    def intersection(self, other):
        # Zadeh intersection (min)
        return {
            x: min(self.mu.get(x, 0), other.mu.get(x, 0))
            for x in set(self.mu) | set(other.mu)
        }


# Example
A = FuzzySet({
    'x1': 0.7,
    'x2': 0.4,
    'x3': 0.9
})

B = FuzzySet({
    'x1': 0.6,
    'x2': 0.8,
    'x3': 0.2
})

print("Fuzzy Set A:", A.mu)
print("Fuzzy Set B:", B.mu)
print("Intersection (A ∩ B):", A.intersection(B))
```

Fuzzy Complement (NOT)

```python
# Fuzzy Complement

class FuzzySet:
    def __init__(self, membership):
        self.mu = membership

    def complement(self):
        return {
            x: 1 - mu
            for x, mu in self.mu.items()
        }


# Example
A = FuzzySet({
    'x1': 0.2,
    'x2': 0.6,
    'x3': 0.9
})

print("Fuzzy Set A:", A.mu)
print("Complement of A:", A.complement())
```

Fuzzy difference operation

```python
# Fuzzy Difference (A - B)

class FuzzySet:
    def __init__(self, membership):
        self.mu = membership

    def difference(self, other):
        """
        Fuzzy difference using:
        A - B = min(mu_A(x), 1 - mu_B(x))
        """
        return {
            x: min(self.mu.get(x, 0), 1 - other.mu.get(x, 0))
            for x in set(self.mu) | set(other.mu)
        }


# Example
A = FuzzySet({
    'x1': 0.7,
    'x2': 0.4,
    'x3': 0.9
})

B = FuzzySet({
    'x1': 0.6,
    'x2': 0.8,
    'x3': 0.2
})

print("Fuzzy Set A:", A.mu)
print("Fuzzy Set B:", B.mu)
print("Difference (A - B):", A.difference(B))
```
