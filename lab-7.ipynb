Stocastic gradient decent

```python
import numpy as np

x = np.array([
    [0, 0],     
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 1, 1])

# Initialize weights and bias
w = np.array([0.2, -0.1], dtype=float)
b = 0.0

eta = 0.1
epochs = 5

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

errors_per_epoch = []

# Training
for epoch in range(epochs):
    total_error = 0

    for i in range(len(x)):
        z = np.dot(w, x[i]) + b
        y_hat = sigmoid(z)

        error = y[i] - y_hat

        # Gradient descent update
        w = w + eta * error * x[i]
        b = b + eta * error

        if abs(error) > 0.5:
            total_error += 1

    errors_per_epoch.append(total_error)

# Final predictions
predictions = []
for i in range(len(x)):
    z = np.dot(w, x[i]) + b
    prob = sigmoid(z)
    predictions.append(1 if prob >= 0.5 else 0)

predictions = np.array(predictions)

print("Errors in each epoch:", errors_per_epoch)
print("Final Weights:", w)
print("Final Bias:", b)
print("Final Predictions:", predictions)
```

Mini Batch Gradient Desent

```python
import numpy as np

x = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 1, 1])

# Initialize weights and bias
w = np.array([0.2, -0.1], dtype=float)
b = 0.0

eta = 0.1
epochs = 5
batch_size = 2

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

errors_per_epoch = []

# Training
for epoch in range(epochs):
    total_error = 0

    # Loop over batches
    for i in range(0, len(x), batch_size):
        x_batch = x[i:i+batch_size]
        y_batch = y[i:i+batch_size]

        dw = np.zeros_like(w)
        db = 0

        # Compute gradients over batch
        for j in range(len(x_batch)):
            z = np.dot(w, x_batch[j]) + b
            y_hat = sigmoid(z)

            error = y_batch[j] - y_hat

            dw += error * x_batch[j]
            db += error

            if abs(error) > 0.5:
                total_error += 1

        # Average gradients
        dw = dw / batch_size
        db = db / batch_size

        # Update weights
        w = w + eta * dw
        b = b + eta * db

    errors_per_epoch.append(total_error)

# Final predictions
predictions = []
for i in range(len(x)):
    z = np.dot(w, x[i]) + b
    prob = sigmoid(z)
    predictions.append(1 if prob >= 0.5 else 0)

predictions = np.array(predictions)

print("Errors in each epoch:", errors_per_epoch)
print("Final Weights:", w)
print("Final Bias:", b)
print("Final Predictions:", predictions)
```
